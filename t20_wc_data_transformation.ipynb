## End to End Data Analysis

```python
import json 
import pandas as pd
```

### 1. Transforming and exporting match summary.

```python
with open('t20_json_files/t20_wc_match_results.json') as f:
    data_match = json.load(f)
data_match
```

```python
## Out data is contained in an array which is further inside in an array having just on element i.e, matchSummary.

df_match = pd.DataFrame(data_match[0]["matchSummary"])
df_match.head()
```

```python
## Checking the shape of dataframe.

df_match.shape
```

```python
## Rename the scorecard as match_id for using it as key to connect with other tables.

df_match.rename({'scorecard': 'match_id'}, axis =1 , inplace = True)
```

```python
df_match.head()
```

```python
df_match.to_csv('csv/match_summary.csv', index = False)
```

### 2. Transforming and exporting batting summary.

```python
## Import batting summary.
with open('t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    
    all_data = []
    for rec in data:
        all_data.extend(rec["battingSummary"])
    
df_batting= pd.DataFrame(all_data) 
df_batting.head()
```

```python
## Let's make a column for out or not_out.

df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
df_batting.head(11)
```

```python
df_batting.drop(columns=["dismissal"],inplace = True)
df_batting.head(11)
```

```python
## Data cleaning.

df_batting["batsmanName"] = df_batting["batsmanName"].apply(lambda x: x.replace('â€',''))
df_batting["batsmanName"] = df_batting["batsmanName"].apply(lambda x: x.replace('xaθ',''))

df_batting.head(10)
```

```python
## To join both tables on match_id, a team1VSteam2 columnn is created with the help of dictionary in df_match

match_id_dict = {}

for index in df_match.index:
    key1 = df_match["team1"][index] + " Vs " + df_match["team2"][index]
    key2 = df_match["team2"][index] + " Vs " + df_match["team1"][index]
    
    match_id_dict[key1] = df_match["match_id"][index]
    match_id_dict[key2] = df_match["match_id"][index]
```

```python
## Mapping the dictionary with match name to get match id's.
df_batting["match_id"]= df_batting["match"].map(match_id_dict)

df_batting.head()
```

```python
## Now we export the tables into csv files and with no index coulumn.

df_batting.to_csv('csv/batting_summary.csv',index = False)
```

### 3. Transforming and exporting bowling summary.

```python
with open('t20_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    
all_data = []
for rec in data:
    all_data.extend(rec["bowlingSummary"])
    
df_bowling= pd.DataFrame(all_data) 
df_bowling.head(11)
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_id_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('csv/bowling_summary.csv', index = False)
```

### 4. Transforming and exporting player information.

```python
with open('t20_json_files/t20_wc_player_info.json') as f:
    data = json.load(f)
```

```python
df_players = pd.DataFrame(data)
df_players.head(10)
```

```python
## Data cleaning.

df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head(22)
```

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('(c)',''))
df_players.head(22)
```

```python
df_players.to_csv('csv/players_without_images.csv', index = False)
```
